### Feature Engineering

```python
age_bins = [0,24,30,100]
bin_labels = ["Young", "Mid", "Senior"]

df['AgeBin'] = pd.cut(df["Age"],bins=age_bins,labels=bin_labels)
```

```python
df
```

```python
salary_bin = [0,50000,75000,100000]
salary_bin_labels = ["low","mid","high"]
df['SalaryBin'] = pd.cut(df['Salary'],bins=salary_bin, labels=salary_bin_labels)
```

## Section 04 - EDA and Visualization Interpretation 

```python
import pandas as pd

data = {
    "Salary": [45000, 52000, 58000, 60000, 61000, 63000, 70000, 72000, 150000],
    "Age":    [22, 25, 27, 30, 32, 35, 40, 45, 29],
    "Experience": [1, 2, 3, 5, 6, 7, 10, 15, 4]
}

df = pd.DataFrame(data)
df
```

### Historgram of Salary 

```python
df.Salary.plot(kind='hist')
```

1. **What does the histogram of Salary show?**

    Sample Answer:
    
    The distribution is right-skewed because one very high salary (150,000) stretches the tail.
    
    Most salaries fall between 45,000 and 75,000.
    
    The presence of a large value at 150,000 suggests a possible outlier.
    
    The central cluster indicates that most employees earn within a normal band.

2. **Are there outliers?**

     Yes.
    
    Salary = 150,000 is far from the other observations  an outlier.

3. **Key insights from the histogram**
    
    Salary is not normally distributed.
    
    Most salaries are in the mid-range (45k75k).
    
    An unusually high salary may indicate:
    
    a senior-level employee
    
    data entry error
    
    special position or bonus

### Boxplot of Age 

```python
df.Age.plot(kind='box')
```

1. **What does the boxplot of Age show?**

    Sample Answer:
    
    Ages are mostly between 27 and 40 (IQR range).
    
    The median age is around 3233.
    
    The distribution looks fairly symmetric  no long whiskers.
    
    No extreme differences between lower and upper halves.

2. **Are there outliers?**

     No clear outliers.
    
    All age values lie within a reasonable range.
    
    No ages fall below Q1  1.5IQR or above Q3 + 1.5IQR.

3. **Key insights from the boxplot**

    The workforce is mostly young to middle-aged adults.
    
    A stable spread of ages  no extreme young or extremely old employees.
    
    Indicates a balanced age distribution.

```python
df.plot(kind='scatter', x='Experience', y='Salary')
```

1. **Does the scatterplot show correlation?**

     Yes  positive correlation.
    
    Explanation:
    
    As Experience increases, Salary tends to increase.
    
    The points roughly follow an upward trend.
    
    This suggests higher experience is associated with higher pay.
    
    BUT
    
    The outlier salary (150,000) makes the relationship slightly distorted.

2. **Are there outliers?**

     Yes.
    
    The point where Salary = 150,000 with Experience = 4 is an outlier.
    
    It does not follow the general trend.

3. **Key insights from the scatterplot**

    More experienced individuals generally earn higher salaries.
    
    Salary growth is not perfectly linear but shows a clear upward direction.
    
    The outlier may represent:
    
    a special role
    
    a performance bonus
    
    incorrect data
