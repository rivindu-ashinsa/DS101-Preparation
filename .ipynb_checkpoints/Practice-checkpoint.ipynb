## Section 03 - Pandas 

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ayesha", "Kamal", "Ravi", "Nimal", "Saman", "Rashmi"],
    "Age": [23, 31, 21, 29, 35, 27],
    "Department": ["Sales", "HR", "Sales", "IT", "IT", "HR"],
    "Salary": [50000, 62000, 48000, 75000, 80000, 59000],
    "Experience": [1.2, 4.5, 0.8, 6.0, 7.2, 3.0],
    "Remote": ["Yes", "No", "Yes", "No", "No", "Yes"]
})
```

```python
df_no_duplicates = df.drop_duplicates()
df_no_duplicates
```
