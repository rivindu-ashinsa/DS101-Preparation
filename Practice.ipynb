## Section 04 - EDA and Visualization Interpretation 

```python
import pandas as pd

data = {
    "Salary": [45000, 52000, 58000, 60000, 61000, 63000, 70000, 72000, 150000],
    "Age":    [22, 25, 27, 30, 32, 35, 40, 45, 29],
    "Experience": [1, 2, 3, 5, 6, 7, 10, 15, 4]
}

df = pd.DataFrame(data)
df
```

### Historgram of Salary 

```python
df.Salary.plot(kind='hist')
```

1. **What does the histogram of Salary show?**

    Sample Answer:
    
    The distribution is right-skewed because one very high salary (150,000) stretches the tail.
    
    Most salaries fall between 45,000 and 75,000.
    
    The presence of a large value at 150,000 suggests a possible outlier.
    
    The central cluster indicates that most employees earn within a normal band.

2. **Are there outliers?**

     Yes.
    
    Salary = 150,000 is far from the other observations  an outlier.

3. **Key insights from the histogram**
    
    Salary is not normally distributed.
    
    Most salaries are in the mid-range (45k75k).
    
    An unusually high salary may indicate:
    
    a senior-level employee
    
    data entry error
    
    special position or bonus

### Normal destribution

```python
import pandas as pd

df = pd.DataFrame({
    "Salary":       [45000,52000,58000,60000,61000,63000,70000,72000,150000,52000,58000,62000,47000,51000],
    "Age":          [22,25,27,30,32,35,40,45,29,31,28,33,26,24],
    "Experience":   [1,2,3,5,6,7,10,15,4,3,2,6,1,1],
    "Gender":       ["Male","Female","Male","Female","Female","Male","Male","Female","Male","Female","Female","Male","Male","Female"],
    "Department":   ["IT","HR","IT","Finance","IT","Finance","HR","IT","Finance","HR","IT","Finance","IT","Finance"],
    "Purchase":     ["Yes","No","Yes","No","Yes","Yes","No","Yes","No","Yes","No","No","Yes","No"],
    "TestScore_Before": [56,60,58,62,55,57,70,75,65,59,61,63,57,58],
    "TestScore_After":  [68,72,70,74,62,66,78,82,72,70,71,75,65,69],
    "DailyCalls":   [3,4,5,2,6,3,4,5,2,3,4,3,6,2],   # for Poisson
    "Conversions":  [1,1,2,0,3,1,1,2,0,1,1,1,3,0],   # for binomial-like tests
})

df
```

```python
df['Age'].plot(kind='hist')
```

```python
scale = df['Age'].std()
loc = df['Age'].mean()

stats.norm.cdf(30,scale=scale, loc=loc) # Cumulative distribution (probability of getting >=x)
stats.norm.pdf(30,scale=scale, loc=loc) # height of the probability distribution curve
stats.norm.cdf(30,scale=scale, loc=loc) # inverse of CDF
```

### ttest (1 sample)

```python
stats.ttest_1samp(df['Age'],popmean=50)
```

### ttest (two groups)

```python
stats.ttest_ind(df['Age'],df['Experience'])  # if p-value is less than 0.05, we can reject null hypothesis (means are equal). 
```

### Paired ttest

```python
# compare values like before and after
stats.ttest_rel(df.TestScore_Before, df.TestScore_After)
```

### Binomial Distribution Probability 

- Binomial distribution probability is the chance of getting a specific number of "successes" in a set number of independent trials, where each trial has only two possible outcomes

```python
import scipy.stats as stats

stats.binom.pmf(k=6, n=10, p=0.5) # here k refers to the number of successes while N refers to the number of independent events
```

### Poisson distribution probability

- A Poisson probability is a discrete probability distribution that calculates the likelihood of a specific number of events (\(k\)) occurring in a fixed interval of time or space, given a known average rate of occurrence (\(\lambda \))

```python
from scipy.stats import poisson

poisson.pmf(k=5, mu=4) # mu refers to the mean 
```

```python
poisson.pmf(k=12, mu=df['Experience'].mean())
```
