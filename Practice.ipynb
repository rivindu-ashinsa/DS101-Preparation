### CHI-Square Test

#### CHI-Square Distribution - Degree 1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats # Import the necessary scipy stats module

# 1. Generate Data
# Generate samples from a standard normal distribution
sample_dis = stats.norm.rvs(loc=0, scale=1, size=20000)

# Calculate the chi-squared distribution (z^2)
# Rounding is not necessary for plotting the histogram
chi_sq_dis = sample_dis ** 2

# 2. Create Figure and Axes using the object-oriented approach (nicer than plt.subplot calls)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# 3. Plot the standard normal distribution histogram
axes[0].hist(sample_dis, bins=50, color='skyblue', edgecolor='black')
axes[0].set_title('Standard Normal Distribution (Z)')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].grid(True, alpha=0.6)

# 4. Plot the chi-squared distribution histogram (1 degree of freedom)
axes[1].hist(chi_sq_dis, bins=50, color='salmon', edgecolor='black', range=[0, 10])
axes[1].set_title(r'Chi-Squared Distribution ($Z^2$)')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')
axes[1].grid(True, alpha=0.6)

# 5. Adjust layout automatically for better spacing
plt.tight_layout()

# You can also manually adjust spacing if needed:
# plt.subplots_adjust(wspace=0.3)

# 6. Display the plots
plt.show()
```

#### CHI-Square Distribution - Degree 2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# --- Configuration ---
NUM_SAMPLES = 20000
DEGREES_OF_FREEDOM = 2

# 1. Generate Data (Standard Normal RVs)
standard_normal_samples = stats.norm.rvs(loc=0, scale=1, size=NUM_SAMPLES)

# 2. Reshape and Calculate Chi-Squared values
sample_pairs = standard_normal_samples.reshape(-1, DEGREES_OF_FREEDOM)
squared_samples = sample_pairs ** 2
chi2_samples_df10 = np.sum(squared_samples, axis=1)

# 3. Create Figure and two Subplots side-by-side
# We request 1 row and 2 columns for the axes layout
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# --- Plot 1: Standard Normal Distribution ---
ax1 = axes[0]
ax1.hist(standard_normal_samples, bins=50, color='skyblue', edgecolor='black', density=True)
# We set density=True here to show probability density, which makes the shape clearer
ax1.set_title('A) Standard Normal Distribution (Z)')
ax1.set_xlabel('Value')
ax1.set_ylabel('Density')
ax1.grid(alpha=0.6)

# --- Plot 2: Chi-Squared Distribution (df=10) ---
ax2 = axes[1]
ax2.hist(chi2_samples_df10, bins=50, color='salmon', edgecolor='black', density=True)
ax2.set_title(f'B) Chi-Squared Distribution (df = {DEGREES_OF_FREEDOM})')
ax2.set_xlabel('Sum of Squared Normal Variables')
ax2.set_ylabel('Density')
ax2.grid(alpha=0.6)

# 4. Adjust layout automatically for better spacing
# This prevents the titles and labels from overlapping
plt.tight_layout(w_pad=2.0) # Using w_pad to ensure enough space between plots

plt.show()
```
